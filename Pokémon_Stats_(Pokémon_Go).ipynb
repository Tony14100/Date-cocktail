Этот набор данных содержит статистику, атрибуты и описания покемонов в Pokémon Go. Используйте этот набор данных, чтобы найти закономерности между различными первичными или вторичными типами, регионами и категориями покемонов.

Данные
число - количество покемонов в Pokédex
pokemon_name - имя покемона
pic_url - URL -адрес PNG-файла покемона
main_type - основной тип покемонов
secondary_type - вторичный тип покемона (или НУЛЕВОЙ только 1 тип)
регион - Регион , где покемоны впервые появились в играх Pokémon
категория - Категория покемонов
рост - рост покемона
вес - Вес покемона
pokemon_family - производное семейство покемонов (напр. Чармандер, Чармелеон, Чаризард будут семьей Чармандера)
атака - статистика атаки базы покемонов в Pokémon Go
защита - статистика защиты базы покемонов в Pokémon Go
выносливость - базовая характеристика выносливости покемонов в Pokémon Go
cp_range - диапазон боевой мощи (CP) покемонов в Pokémon Go
hp_range - диапазон хит-пойнта (CP) покемонов в Pokémon Go
capture_rate - скорость захвата покемонов в Pokémon Go (значения 'N / A' указывают на то, что данные не были доступны из производного источника данных)
скорость бегства - скорость бегства покемонов в Pokémon Go (значения 'N / A' указывают на то, что данные не были доступны из производного источника данных)
male_perc - процент найденных покемонов мужского пола (значения 'N / A' указывают, что покемоны бесполые, только мужчины или только женщины)
female_perc - процент найденных покемонов женского пола (значения 'N / A' указывают, что покемоны бесполые, только мужчины или только женщины)
сопротивление - словарь сопротивлений покемонов определенным типам в следующей форме: {'percent_resistant': ['type_1', 'type_2']}
слабости - словарь слабых сторон покемонов для определенных типов в следующей форме: {'процент_весомости': ['тип_1', 'тип_2']}
wild_avail - доступен в дикой природе?
egg_avail - Доступно из яиц?
raid_avail- доступен в качестве рейдового босса?
research_avail - доступно из полевых исследований?
доступна ли блестящая форма?
тень - доступна форма тени?
pkedex_desc - Pokémon's Pokédex description
poss_attacks - список возможных атак, доступных для покемонов

https://pokemongo.fandom.com/
